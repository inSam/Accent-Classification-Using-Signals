```python
import pandas as pd
import librosa
import os
import numpy as np
```

```python
df = pd.read_csv('speakers_all.csv')
df = df[df['file_missing?'] == False].reset_index()
df
```

```python
df.groupby('native_language')['native_language'].count() \
                                                .reset_index(name='count') \
                                                .sort_values(['count'], ascending=False)\
                                                .tail(9)                                           
```

```python
def buildset(country_list, df, path):
    X = []
    y = []
    label = 0
    for country in country_list:
        filenames = df[df['native_language'] == country]['filename']
        for name in filenames:
            data, sr = librosa.load(os.path.join(path, name+'.mp3'))
            X.append(data)
            y.append(label)
        label += 1
    return X, np.array(y)
```

```python
X, y = buildset(['dutch', 'russian'], df, 'recordings/recordings')
```
